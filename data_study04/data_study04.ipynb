```python
df1.head(2)
```

```python
#amount(금액) 가장 높은 10개의 항목을 가지고 오고 싶다. 
p1 = pd.pivot_table(data = df1, index = 'channel', values = 'amount', aggfunc = 'sum')
p1.sort_values(by = 'amount', ascending = False).head(10)
top_list = p1.sort_values(by = 'amount', ascending = False).head(10).index.tolist()
```

```python
df1['channel'].value_counts()
```

* 채널이 많다고 해서 꼭 많이 팔리는 것은 아니라는 것을 알 수 있다. 

```python
sns.barplot(data = df1, x = 'channel', y = 'amount',order = top_list)
```

* 그래프는 매출의 평균을 나타내고 있다. 

```python
# 총 amount을 구하고 싶다. 
sns.barplot(data = df1, x = 'channel', y = 'amount',order = top_list, estimator = sum)
```

* 그래프의 검은 점은 무엇일까?
A : 신뢰구간이라는 값이 구해져서 나타내지고 있다. 

```python
# 검은 점을 없애고 싶다. 
sns.barplot(data = df1, x = 'channel', y = 'amount',order = top_list, estimator = sum, ci = None)
```

```python
# bar plot또한 overlay가 가능하다. 
sns.barplot(data = df1, x = 'channel', y = 'amount',order = top_list,
            estimator = sum, ci = None, hue = 'type_of_contract')
```

```python
# boxplot은 데이터의 분포를 박스 형태로 출력해준다. 
sns.boxplot(data = df1, x = 'channel', y = 'amount', order = top_list)
```

* 제일 밑 작대기 : 최저점
* 제일 위 작대기 : 최고점
* 박스 제일 밑 작대기 : 1사분위 점(25%)
* 박스 중간 작대기 : 2사분위 점(50%) == 평균
* 박스 제일 위 작대기 : 3사분위 점(75%)
* 그 외 점들 : 신뢰구간을 벗어나는 이상치이다.

### 다변수
 - 순서를 가지는 데이터에 대한 그래프를 한 번 그려보자

```python
df1['datetime(timestmp)'] = pd.to_datetime(df1['datetime'])
```

```python
sns.lineplot(data = df1, x ='datetime(timestmp)', y = 'amount')
```

* 위의 그래프는 일자별로 평균의 amount값을 계산하여 그래프에 나타내었다.

```python
# 만약 내가 사장이라면 하루 평균 매출이 아닌 총 하루 매출량이 더욱 궁금할 것이다.
# 그래프 옆에 희미한 표시 -> 평균에 따른 신뢰구간도 구해서 그래프에 나타내어졌다.
```

```python
sns.lineplot(data=df1, x= 'datetime(timestmp)', y = 'amount',
             ci =None, estimator = sum)
```

* 총량으로 했음에도 불구하고 삐죽삐죽한 모양의 그래프가 나왔다. 왜 `그럴까?

```python
df1['계약요일'] = df1['datetime(timestmp)'].dt.day_name()
```

```python
sns.barplot(data = df1, x = '계약요일', y = 'amount', estimator = sum, ci = None)
```

* 위의 그래프에서 보면 알 수 있듯이 토요일날 매출이 주기적으로 낮은 것을 볼 수 있다. 

* 따라서 요일, 날짜별로 amount값을 보는 것이 아니라 주차나 월별 amount값을 보면
이를 해결 할 수 있다. 

```python
df1['계약주차'] = df1['datetime(timestmp)'].dt.isocalendar().week
```

```python
sns.lineplot(data=df1, x = '계약주차', y = 'amount', ci = None, estimator = sum)
```

* 왜 이런 데이터가 나오는 것일까??

```python
df1['datetime(timestmp)'].describe()
```

* week라는 함수는 해당년도 1년을 기준으로 1~52까지 숫자를 매긴다. 
해당 표는 19년도 10월부터 20년도 2월까지 밖에 없으므로 
빈 주차들을 채워 넣을 수 없다. 

```python
df1['계약연도'] = df1['datetime(timestmp)'].dt.year
```

```python
sns.pointplot(data = df1, x = '계약주차', y = 'amount', ci = None,
              estimator = sum, hue = '계약연도')
```

* 이럴 경우 이상치인 1,6,42 주차는 없앤다고 하면 좋은 데이터가 될 수 있다. 

```python
#2020년과 2019년을 나눠서 분석하는 방법
cond1 = (df1["계약연도"] == 2019) 
df2 = df1.loc[cond1]
sns.pointplot(data = df2, x = '계약주차', y = 'amount', ci = None,
              estimator = sum, hue = '계약연도')
```
