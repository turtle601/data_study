```python
# 모든 범주형 데이터들의 describe를 알고 싶을 때

# 전체 column 목록 - 숫자형 column 목록을 빼면 된다. 

total_list = df1.columns.tolist()
numeric_list = df1.describe().columns.tolist()

object_list = list(set(total_list) - set(numeric_list))
```

```python
df1[object_list].describe()
```

```python
# 정렬

df1.sort_values(by = '입고수량') #오름차순
```

```python
df1.sort_values(by = '입고수량', ascending = False) #내림차순
```

```python
# 여러 개의 항목을 정렬할 때

df1.sort_values(by = ['입고수량','출고수량'])
```

* 입고수량이 오름차순으로 정렬, 같은 입고수량일 때 출고수량도 오름차순으로 정렬

```python
df1.sort_values(by = ['입고수량','출고수량'],ascending = [True,False])
```

* 입고수량이 오름차순으로 정렬, 같은 입고수량일 때 출고수량은 내림차순으로 정렬
