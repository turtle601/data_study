### 단일변수 시각화

```python
import seaborn as sns
```

```python
 #.describe의 항목에 따른 count의 값을 알고 싶을 때
df1['type_of_contract'].value_counts()
```

```python
sns.countplot(data=df1, x = 'product', hue = 'type_of_contract2')
```

* product에 따라 type_of_contract2 항목 또한 나누어서 표시하였다. 

```python
import matplotlib.pyplot as plt #그래프 출력과 직접적 연관이 있는 
import matplotlib as mpl        #한글 폰트 등의 옵션 등을 잡아준다. 
```

```python
# 그래프 출력 시 한글 깨짐 방지
mpl.rc("font", family = 'Malgun Gothic')
```

```python
sns.countplot(data = df1, x = 'type_of_contract')
```

```python
#그래프 사이즈 늘리기 - 앞서 범례와 그래프를 구분하기 위해서
plt.figure(figsize = [10,5])
sns.countplot(data=df1, x = 'product', hue = 'type_of_contract2')

#범례의 위치를 오른쪽으로 옮김
plt.legend(loc='right') 
```

### 연속형 변수

```python
#연속된 변수를 확인해보고 싶다. - 히스토그램
sns.histplot(data = df1, x = 'age')
```

```python
# 확률 분포라는 선이 생성
plt.title('계약 유형 별, 고객 연령 분포')
sns.histplot(data = df1, x = 'age',kde = True, hue  = 'type_of_contract')
```

* 이러한 그래프로 이상치도 확인 가능하다. 

```python
sns.countplot(data = df1, x = 'bank')
```

* bank 목록이 많아서 보기 싫다.

```python
plt.figure(figsize = [5,10])
sns.countplot(data=df1, y = 'bank')
```

* 그래프에서 볼 수 있듯이 bank라는 항목에 편차가 조금 있다. 따라서 특정 값만 빼와서 데이터를 시각화 하고 싶다.

```python
# df1['bank'].value_counts() 를 하면 count가 높은 순서부터 낮은 순서까지 차례로 볼 수 있다. 
sns.countplot(data=df1, x = 'bank', order = ['국민은행','롯데카드','농협은행','신한은행'])
```

```python
order_list = df1['bank'].value_counts().index.tolist()
plt.figure(figsize = [10,5])
sns.countplot(data = df1, x= 'bank', order = order_list[:10]) #내림차순 0~10까지 보고 싶다. 
```
